##### ***`Personas`***

- `Family` – “The Family segment have relatively high discretionary spend, but they have to make it stretch to meet the needs of the whole family and for that reason they appreciate a good deal or a discount. They use their card almost everyday across all categories, not just essential. They enjoy spending on family occasions and outings with the children. These working families are more likely to be urban, have pets – and a car is required to transport them all. They are health conscious and tend to transact during lunch or in the afternoon.”
 
- `Frugal Family` – “Frugal families have average discretionary spend, but lower than average for a family. For that reason, these individuals spend more across the budget merchants, which is why their average transaction value is also lower. They tend to live in cities, can afford to run a car, but not much is left over so spend is concentrated in essential categories. Despite them being on a budget, they enjoy spending on special occasions.”
 
- `Affluent Spenders` – “Affluent segment have much higher discretionary spend and have high spend in affluent brands. They enjoy shopping at department stores and traveling internationally. These cardholders are more likely to shop in the morning or at lunchtime and their spend is centred on personal services such as hairdressers and massage therapists.”
 
- `Urban Spenders` – “Urban Spenders have the highest discretionary spend and tend to be acquiring more responsibility in their lives and the world. They still enjoy socialising, but are more likely to be settled, own a car and have pets. They look after themselves by going to the gym in their free time. They enjoy celebrating special occasions.”
 
- `Sensible Spenders` – “Sensible Spenders have a much lower level of spend, but a relatively high average transaction value, suggesting infrequent treats for themselves. When they do shop, it is not across the budget brands. They are more likely to have a car and pets, but generally keep their regular spend to essential categories. They prefer face to face experience rather than shopping online.”

##### ***`Card Data`***

| Attribute             | Description                                                                 |
|-----------------------|-----------------------------------------------------------------------------|
| `cpd_mnth_id`         | The month and year of the transaction in the format YYYYMM.                 |
| `cpd_dt`              | The date of the transaction.                                                |
| `cluster_name_adjusted` | The adjusted name of the customer segment or cluster.                     |
| `mrch_catg_cd`        | The merchant category code.                                                 |
| `mrch_catg_rlup_nm`   | The rolled-up name of the merchant category.                                |
| `merchant`            | The name of the merchant.                                                   |
| `cp_flag`             | A flag indicating whether the transaction was a card-present transaction (1) or not (0). |
| `domestic_flag`       | A flag indicating whether the transaction was domestic (1) or not (0).      |

| `interregion_flag`    | A flag indicating whether the transaction was between different regions (1) or not (0). |
| `city_name`           | The name of the city where the transaction took place.                      |

| `spend`               | The amount spent in the transaction.                                        |

```python
# Function created to style tables
def style(data):
    # Create a GreatTable object and set basic options
    table = gt.GT(data).tab_options(
        table_width='100%',
        table_background_color="#234a8c",
        table_font_color="black",
        table_font_names=['PT Mono'],
        table_font_color_light='white',
        column_labels_font_weight='bold',
        column_labels_background_color="#efaa43").cols_align(align="left", columns=None)

    return table
```

#### ***`Cluster Analysis`***

```python
# Define a custom function to calculate mode, count of visits
def mode_and_count(x):
    mode = x.mode()[0] if len(x.mode()) > 0 else None  # Mode of 'merchant'
    count = x.value_counts().max()  # Count of times the mode merchant was visited
    return pd.Series([mode, count], index=['fav_merchant', 'transaction_count'])

# Group by 'cluster_name_adjusted' and calculate aggregations
groupby_cluster = groceries.groupby('cluster_name_adjusted').agg(
    merchant_mode_and_count=('merchant', mode_and_count),  # Apply custom function
    overall_average_spend=('spend', 'mean'),               # Overall mean of 'spend'
    overall_total_spent=('spend', 'sum'),
    total_transactions=('spend', 'count')# Overall sum of 'spend'
).reset_index()

# Extract 'merchant' mode and visit count into separate columns
groupby_cluster[['fav_merchant', 'transaction_count']] = groupby_cluster['merchant_mode_and_count'].apply(pd.Series)

# Calculate average and total spend for the favorite merchant
fav_merchant_stats = groceries.groupby(['cluster_name_adjusted', 'merchant']).agg(
    average_spend=('spend', 'mean'),
    total_spent=('spend', 'sum')
).reset_index()

# Merge favorite merchant stats back into the main dataframe
groupby_cluster = groupby_cluster.merge(
    fav_merchant_stats,
    how='left',
    left_on=['cluster_name_adjusted', 'fav_merchant'],
    right_on=['cluster_name_adjusted', 'merchant']
)

# Drop the intermediate columns and rename for clarity
groupby_cluster.drop(columns=['merchant_mode_and_count', 'merchant'], inplace=True)
groupby_cluster.rename(columns={
    'average_spend': 'fav_merchant_average_spend',
    'total_spent': 'fav_merchant_total_spent',
    'transaction_count':'fav_merchant_transactions',
    'overall_average_spend': 'average_spend_overall',
    'overall_total_spent': 'total_spent_overall'
}, inplace=True)

# Adjust for better reading
groupby_cluster = groupby_cluster[[
    'cluster_name_adjusted',
    'fav_merchant',
    'fav_merchant_transactions',
    'fav_merchant_average_spend',
    'fav_merchant_total_spent',
    'total_transactions',
    'average_spend_overall',
    'total_spent_overall'
]]
```

```python
style(groupby_cluster)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(24, 6))

# Plot for transaction count
sns.barplot(data=groupby_cluster, x='cluster_name_adjusted', y='total_transactions', palette='tab10', ax=ax[0], edgecolor='black')
ax[0].set_title('Transaction Count by Cluster', fontsize=15)
ax[0].set_xlabel('Cluster Name')
ax[0].set_ylabel('Transaction Count')
ax[0].grid(True, color='white')

# Plot for average spend
sns.barplot(data=groupby_cluster, x='cluster_name_adjusted', y='average_spend_overall', palette='tab10', ax=ax[1], edgecolor='black')
ax[1].set_title('Average Spend by Cluster', fontsize=15)
ax[1].set_xlabel('Cluster Name')
ax[1].set_ylabel('Average Spend (£)')
ax[1].grid(True, color='white')

# Plot for average spend
sns.barplot(data=groupby_cluster, x='cluster_name_adjusted', y='total_spent_overall', palette='tab10', ax=ax[2], edgecolor='black')
ax[2].set_title('Total Spent by Cluster', fontsize=15)
ax[2].set_xlabel('Cluster Name')
ax[2].set_ylabel('Total Spend (£)')
ax[2].grid(True, color='white')

plt.show()
```

#### ***`CO`<sub>`2`</sub> `Footprint Per Transaction`***

This value would ideally be calculated using the Ecolytiq API, but i have been having trouble accessing it. However , through other sources i was able to obtain some information, that can help in finding out the CO<sub>2</sub> footprint per transaction :


1. **Average Carbon Footprint**: The average carbon footprint for a person in the UK is around 6 tonnes of CO2 equivalent (CO2e) per year. <a href='https://www.theecoexperts.co.uk/blog/average-carbon-footprint-uk' alt='link'>Source</a>

2. **Economic Data**: According to the UK's Office for National Statistics, the UK's GDP was approximately £2.3 trillion in 2020. <a href='https://www.ons.gov.uk/economy/environmentalaccounts/bulletins/ukenvironmentalaccounts/2022' alt='link'>Source</a>

Using these figures, we can estimate the carbon footprint per pound of GDP. Here's a rough calculation:

- Total UK carbon footprint (6 tonnes per person × ~67 million people) = 402 million tonnes of CO2e per year.
- UK's GDP = £2.3 trillion per year.

Thus, carbon footprint per £ of GDP:

***Total UK Carbon Footprint / UK GDP ≈ 402 million tonnes CO<sub>2</sub>e / £2.3 trillion ≈ 0.175 kg CO<sub>2</sub>e per £***

`This rough estimate suggests that for every pound spent or earned in the UK economy, approximately 0.175 kg of CO2e is emitted`.

These figures provide a general idea but may vary based on more detailed sector-specific data and other variables. For more precise and sector-specific data, accessing detailed reports from governmental or specialized environmental data sources would be necessary.

```python
# CO2 footprint per £ spend
CO2_pound = 0.175

# Calculate for each transaction
groceries['CO2_footprint_transaction'] = groceries['spend'] * CO2_pound

style(groceries.head())
```
